```python
import pandas as pd
import numpy as  np   
```

```python
df=pd.read_csv(r"C:\Users\Lenova\Desktop\Luxury housig bangalore\Luxury_Housing_Bangalore.csv")
print(df)
```

```python
df.head()
df.shape
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
df.columns
```

```python
for  col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(df[col].mean())
    else:
        df[col] = df[col].fillna('UnKnown')
```

```python
df= df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
for col in df.columns:
    if 'price' in col or 'cost' in col:
     df[col] = pd.to_numeric(df[col], errors='coerce')

for col in df.columns:
    if 'date' in col:
     df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.head()      
df.shape        
df.info()       
```

```python
clean_path = (r"C:\Users\Lenova\Desktop\Luxury housig bangalore\Luxury_Housing_Bangalore_Cleaned.csv")
df.to_csv(clean_path, index=False)
print("Cleaned data saved at:", clean_path)
```
